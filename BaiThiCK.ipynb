# **Câu 1: Phép toán cơ bản trên vector và ma trận**

```python
from sympy import *
import numpy as np

# a)
A = Matrix([3, -2, 1, 4])
B = Matrix([-1, 5, 0, 2])

sum_AB = A + B
linear_comb = 2*A + 3*B

print("1a) Tổng A + B:", sum_AB.T)
print("   2A + 3B:", linear_comb.T)

# b)
dot_product = A.dot(B)
magnitude_A = sqrt(A.dot(A))
magnitude_B = sqrt(B.dot(B))
angle_rad = acos(dot_product / (magnitude_A * magnitude_B))
angle_deg = angle_rad * 180 / pi.evalf()

print("\n1b) Tích vô hướng A.B:", dot_product)
print("   Góc giữa A và B (độ):", angle_deg.evalf(4))

# c)
C = Matrix([[2, -1, 3], [1, 4, -2], [0, 2, 5]])
CCt = C * C.T

print("\n1c) C * C^T:")
pprint(CCt)
```

# **Câu 2: Nhân ma trận, định thức và hạng ma trận**

```python
# a)
M = Matrix([[2, -1, 3, 0], [1, 4, -2, 5], [0, 2, 6, -3], [-1, 0, 1, 4]])
N = Matrix([[1, 3, -2, 0], [4, -1, 5, 2], [0, 2, 3, -1], [-3, 1, 0, 4]])

MN = M * N
print("\n2a) M × N:")
pprint(MN)

# b)
try:
    det_M = M.det()
    print("\n2b) Định thức của M:", det_M)
except:
    print("\n2b) Không thể tính định thức vì ma trận suy biến")

# c)
rank_N = N.rank()
print("\n2c) Hạng của ma trận N:", rank_N)
```

# **Câu 3: Nghịch đảo ma trận và giải hệ phương trình**

```python
# a)
P = Matrix([[3, -1, 0, 2], [1, 4, -2, 0], [0, 1, 5, -1], [2, 0, 1, 3]])

try:
    P_inv = P.inv()
    print("\n3a) Ma trận nghịch đảo của P:")
    pprint(P_inv)
except:
    print("\n3a) Ma trận P không khả nghịch")

# b)
b = Matrix([6, 8, 3, 7])
solution = P_inv * b
print("\n3b) Nghiệm hệ phương trình:")
pprint(solution)

# c)
verification = P * solution
print("\n3c) Kiểm tra nghiệm (P*x):")
pprint(verification)
print("   Vector hằng số b:", b.T)
```

# **Câu 4: Giá trị riêng, vector riêng và chuẩn hóa**

```python
# a)
Q = Matrix([[5, -2, 1], [0, 3, -1], [2, 0, 4]])
eigenvals = Q.eigenvals()
print("\n4a) Giá trị riêng của Q:")
pprint(eigenvals)

# b)
eigenvects = Q.eigenvects()
print("\n4b) Vector riêng của Q:")
for vec in eigenvects:
    print(f"Giá trị riêng {vec[0]}:")
    for v in vec[2]:
        pprint(v)

# c)
print("\n4c) Vector riêng chuẩn hóa:")
for vec in eigenvects:
    print(f"Giá trị riêng {vec[0]}:")
    for v in vec[2]:
        normalized = v / sqrt(v.dot(v))
        pprint(normalized)
```

# **Câu 5: Ứng dụng thực tế**

```python
import numpy as np
import sympy as sp
from sympy import Matrix, pprint

# Dữ liệu đầu vào
R_numpy = np.array([[4, -1, 2], [1, 3, -2], [0, 2, 5]])
b_numpy = np.array([12, 5, 9])

# a) Giải hệ phương trình Rx = b dùng np.linalg.solve()
x_numpy = np.linalg.solve(R_numpy, b_numpy)
print("5a)Nghiệm của hệ phương trình Rx = b (dùng NumPy):")
print("x =", x_numpy)

# b) Dùng SymPy để kiểm tra khả nghịch và tính lại nghiệm bằng nghịch đảo
R_sym = Matrix(R_numpy)
b_sym = Matrix(b_numpy)

# Kiểm tra khả nghịch
if R_sym.det() != 0:
    print("\n5b)Ma trận R khả nghịch. Nghịch đảo R là:")
    R_inv = R_sym.inv()
    pprint(R_inv)

    # Giải lại hệ bằng cách nhân R⁻¹ * b
    x_sym = R_inv * b_sym
    print("\nNghiệm hệ phương trình (dùng R⁻¹ * b):")
    pprint(x_sym)

    # So sánh 2 nghiệm
    print("\nSo sánh nghiệm từ NumPy và SymPy:")
    print("NumPy:", np.round(x_numpy, 4))
    print("SymPy:", [sp.N(val, 4) for val in x_sym])
else:
    print("Ma trận R không khả nghịch.")

# c) Tính chuẩn Frobenius của ma trận R
fro_norm = np.linalg.norm(R_numpy, 'fro')
print(f"\n5c)Chuẩn Frobenius của ma trận R: {fro_norm:.4f}")
```

#**Bài 1**

```python
#Bài 1
#Hàm “scale” tỉ số vector bằng một giá trị thực
def scale(a, v):
  return [a*vi for vi in v]
v = [3,5,7]
print('Hàm “scale” tỉ số vector bằng một giá trị thực',scale(10, v))
#Hàm lấy tổng hai vector
def sumvector(v, w):
  return [vi+wi for (vi, wi) in zip(v, w)]
v = [3,5,7]
w = [2,4,6]
print('Hàm lấy tổng hai vector',sumvector(v, w))
#Hàm nhân 2 vector vô hướng
def dotvector(v, w):
  return sum([vi*wi for (vi, wi) in zip(v, w)])
print('Hàm nhân 2 vector vô hướng',dotvector(v, w))
#Hàm tính chiều dài một vector
def lenvector(v):
  return dotvector(v,v)
print('Hàm tính chiều dài một vector',lenvector(w))
```

```python
#Problem 1: Giao điểm hai đường thẳng trong R^2
import numpy as np
a1 = np.array([[1, -1],
               [2, 3]])
b1 = np.array([-2, 6])
x1 = np.linalg.solve(a1, b1)
print("Kết quả của Problem 1 (x, y):", x1)
```

```python
# Problem 2: Giao điểm ba mặt phẳng trong R^3
a2 = np.array([[1, -1, 0],
               [2, -1, -1],
               [1, 1, 1]])
b2 = np.array([2, 3, 6])
x2 = np.linalg.solve(a2, b2)
print("Kết quả của Problem 2 (x, y):", x2)
```

```python
# Problem 3: Tìm hệ số đa thức bậc hai p(x) = ax^2 + bx + c
a3 = np.array([[1, 1, 1],
               [4, 2, 1],
               [9, 3, 1]])
b3 = np.array([4, 3, 4])
x3 = np.linalg.solve(a3, b3)
print("Kết quả của Problem 3 (x, y):", x3)
```

```python
# Problem 4: Phân rã thành phân số đơn giản để tính tích phân
a4 = np.array([[1, 0, 1],
               [1, 1, -2],
               [-2, 2, 1]])
b4 = np.array([1, -3, 0])
x4 = np.linalg.solve(a4, b4)
print("Kết quả của Problem 4 (x, y):", x4)
```

# **Câu 2**

```python
# Import, khởi tạo biến
from sympy import symbols, Eq, solve, Matrix
x, y, z = symbols('x y z')
a, b, c = symbols('a b c')
```

```python
#Problem 1: Giao điểm hai đường thẳng trong R^2
eq1 = Eq(x - y, -2)
eq2 = Eq(2*x + 3*y, 6)
sol1 = solve((eq1, eq2), (x, y))
print('Problem 1: ', sol1)
```

```python
#Problem 2: Giao điểm ba mặt phẳng trong R^3
eq3 = Eq(x - y, 2)
eq4 = Eq(2*x - y - z, 3)
eq5 = Eq(x + y + z, 6)
sol2 = solve((eq3, eq4, eq5), (x, y, z))
print('Problem 2: ', sol2)
```

```python
# Problem 3: Tìm hệ số đa thức bậc hai p(x) = ax^2 + bx + c
# p(1) = 4, p(2) = 3, p(3) = 4
# p(1) = 4:
eq6 = Eq(a + b + c, 4)
# p(2) = 3:
eq7 = Eq(4*a + 2*b + c, 3)
# p(3) = 4:
eq8 = Eq(9*a + 3*b + c, 4)
sol3 = solve((eq6, eq7, eq8), (a, b, c))
print('Problem 3: ', sol3)
```

```python
# Problem 4: Phân rã thành phân số đơn giản để tính tích phân
eq9  = Eq(a + c, 1)
eq10 = Eq(a + b - 2*c, -3)
eq11 = Eq(-2*a + 2*b + c, 0)
sol4 = solve((eq9, eq10, eq11), (a, b, c))
print('problem 4: ', sol4)
```
